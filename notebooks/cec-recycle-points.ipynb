```python
import folium;
import json;

geoJson=json.load(open('../data/recycle-points.geojson'))

campus_map = folium.Map(location=[55.9450167, -3.1894458],
                    zoom_start=13)

folium.GeoJson(geoJson,
               name='geojson'
              ).add_to(campus_map)

folium.LayerControl().add_to(campus_map)
campus_map
```
